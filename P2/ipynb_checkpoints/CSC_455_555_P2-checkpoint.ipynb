```python
import pandas as pd
import numpy as np
import scipy.stats as ss
import random
import mesa

# For a jupyter notebook add the following line:
%matplotlib inline

# The below is needed for both notebooks and scripts
import matplotlib.pyplot as plt
```

```python
# Feel free to USE any code you have learned from the tutorial. 
class Agent(mesa.Agent):
    """Agents in the CovidModel class below"""
    
    def __init__(self, unique_id, model):
        super().__init__(unique_id, model)
        """
        Add in additional property to the agent
        Hint: for the infection rate, bool(ss.bernoulli.rvs( INFECTION_RATE )) would be helpful.
        Ex. 
            self.wealth = 0
        """
        #CODE HERE
        self.immune = False
        self.infected = False
        self.recovery_time = 0
        self.immunity_time = 0
        
    def move(self):
        possible_steps = self.model.grid.get_neighborhood(
            self.pos, moore=True, include_center=False
        )
        new_position = self.random.choice(possible_steps)
        self.model.grid.move_agent(self, new_position)
    
    """
    Define additional functions for your agents below here.
    Ex. give_money(self) from the tutorial
    """
    
    def infect(self):
        cellmates = self.model.grid.get_cell_list_contents([self.pos])
        if len(cellmates) > 1:
            for other in cellmates:
                if other == self: # Can't infect yourself
                    continue
                if other.infected: # Don't restart the timer for those already sick
                    continue
                if other.immune: # Cant infect an immune person
                    continue
                other.infected = not other.immune and bool(ss.bernoulli.rvs( self.model.infection_rate ))
                if other.infected:
                    other.recovery_time = self.model.recovery_time
    
    
    def step(self):
        self.move()
        """
        Define your logic here for each step/iteration of your simulation below here.
        
        Ex.
        if self.wealth > 0:
            self.give_money()
        """
        if self.infected and self.recovery_time <= 1:
            self.infected = False
            self.immune = True
            self.recovery_time = 0   
            self.immunity_time = self.model.immunity_time
        elif self.infected:
            self.infect()
            self.recovery_time -= 1
        elif self.immune and self.immunity_time <= 1:
            self.immune = False
            self.immunity_time = 0
        elif self.immune:
            self.immunity_time -= 1
            
            
        
    
class CovidModel(mesa.Model):
    """A COVID model with a number of agents that can potentially transmit COVID
    when they are on the same cell of the grid"""

    def __init__(self, no_agents, width, height, infection_rate, initial_infection, recovery_time, immunity_time):
        self.no_agents = no_agents
        self.grid = mesa.space.MultiGrid(width, height, True)
        self.schedule = mesa.time.RandomActivation(self)
        self.running = True
        """
        Add in additional property to track in the model or set by the user
        Hint: Here would be good to set initial infection rate and perhaps track infection rate
        Also, make sure to update the def __init__  to include those properties.
        Ex. 
            def __init__(self, no_agents, width, height, infection_rate)
        """
        #CODE HERE
        self.infection_rate = infection_rate
        self.initial_infection = initial_infection
        self.recovery_time = recovery_time
        self.immunity_time = immunity_time

        # Create agents and add it the schedule
        initially_infected = round(self.no_agents * self.initial_infection)
        for i in range(initially_infected): # Place initially infected agents
            a = Agent(i, self)
            a.infected = True
            a.recovery_time = recovery_time
            self.schedule.add(a)

            # Add the agent to a random grid cell
            x = self.random.randrange(self.grid.width)
            y = self.random.randrange(self.grid.height)
            self.grid.place_agent(a, (x, y))
        
        for i in range(no_agents, initially_infected, -1): # Place initially healthy agents
            a = Agent(i, self)
            self.schedule.add(a)

            # Add the agent to a random grid cell
            x = self.random.randrange(self.grid.width)
            y = self.random.randrange(self.grid.height)
            self.grid.place_agent(a, (x, y))

        """
        Add in to the datacollector properties you want to track over time in the graph.
        
        Ex. 
            self.datacollector = DataCollector({'Agents with Money': 'richAgents',....)
        """
        self.datacollector = mesa.DataCollector({"Infected Agents": "infectedAgents", "Immune Agents": "immuneAgents"})
        
    """
    Define any properties here that can be used for the DataCollector.
    Hint, number of infected agents, immune are pretty interesting.
    Ex. 
        @property
        def richAgents(self):
            agents = self.schedule.agents
            agents_with_money = [a.money_balance for a in agents]
            return int(np.sum(agents_with_money)))
    """
    @property
    def infectedAgents(self):
        agents = self.schedule.agents
        infected_agents = [a.infected for a in agents]
        return np.sum(infected_agents)
    
    @property
    def immuneAgents(self):
        agents = self.schedule.agents
        immune_agents = [a.immune for a in agents]
        return np.sum(immune_agents)


    def step(self):
        self.datacollector.collect(self)
        self.schedule.step()
```

```python
"""
Task # 9
It is generally more difficult to get to 100% infected when using masks in this scenario.
While some of the above charts show that the masks helped only marginally, this could simply be due to a lack of 
repeated testing. As the percentage of a population wearing masks increased, the peak of infections generally
lowered.
Generally, the larger the initial infection rate, the more people got infected. However, the most effective
way to ensure everyone got infected was, similar to the pre-mask state, implement a high recovery time and a low
immunity time. This insured that people got sick often and remained so for a prolonged period of time.
"""
```

```python
class ModifiedMaskAgent(mesa.Agent):
    """Agents in the CovidModel class below"""
    
    def __init__(self, unique_id, model):
        super().__init__(unique_id, model)
        """
        Add in additional property to the agent
        Hint: for the infection rate, bool(ss.bernoulli.rvs( INFECTION_RATE )) would be helpful.
        Ex. 
            self.wealth = 0
        """
        #CODE HERE
        self.immune = False
        self.infected = False
        self.recovery_time = 0
        self.immunity_time = 0
        self.masked = False
        self.dead = False # This adds a mortality to the disease
        
    def move(self):
        possible_steps = self.model.grid.get_neighborhood(
            self.pos, moore=True, include_center=False
        )
        new_position = self.random.choice(possible_steps)
        self.model.grid.move_agent(self, new_position)
    
    """
    Define additional functions for your agents below here.
    Ex. give_money(self) from the tutorial
    """
    
    def infect(self):
        cellmates = self.model.grid.get_cell_list_contents([self.pos])
        if len(cellmates) > 1:
            for other in cellmates:
                if other == self: # Can't infect yourself
                    continue
                if other.infected: # Don't restart the timer for those already sick
                    continue
                # Can't infect someone who is currently immune
                if other.immune:
                    continue
                if other.dead:
                    continue
                
                # I am assuming naively here that the masks compound in effectiveness if both parties wear them.
                if other.masked and self.masked:
                    other.infected = bool(ss.bernoulli.rvs((self.model.masked_infection_rate)**2))
                elif other.masked or self.masked:
                    other.infected = bool(ss.bernoulli.rvs(self.model.masked_infection_rate))
                else:
                    other.infected = bool(ss.bernoulli.rvs(self.model.infection_rate))
                if other.infected:
                    other.recovery_time = self.model.recovery_time
    
    def step(self):
        
        if self.dead:
            return
        
        self.dead = self.infected and bool(ss.bernoulli.rvs(self.model.death_rate)) # Possibility of dying to the disease
        
        if self.dead: # Can't move if you're dead.
            self.infected = False
            self.immune = False
            return
        self.move()
        """
        Define your logic here for each step/iteration of your simulation below here.
        
        Ex.
        if self.wealth > 0:
            self.give_money()
        """
        if self.infected and self.recovery_time <= 1:
            self.infected = False
            self.immune = True
            self.recovery_time = 0   
            self.immunity_time = self.model.immunity_time
        elif self.infected:
            self.infect()
            self.recovery_time -= 1
        elif self.immune and self.immunity_time <= 1:
            self.immune = False
            self.immunity_time = 0
        elif self.immune:
            self.immunity_time -= 1
            
            
        
    
class ModifiedMaskCovidModel(mesa.Model):
    """A COVID model with a number of agents that can potentially transmit COVID
    when they are on the same cell of the grid"""

    def __init__(self, no_agents, width, height, infection_rate, initial_infection, recovery_time, immunity_time,
                 mask_wearing_percentage, masked_infection_rate, death_rate):
        self.no_agents = no_agents
        self.grid = mesa.space.MultiGrid(width, height, True)
        self.schedule = mesa.time.RandomActivation(self)
        self.running = True
        """
        Add in additional property to track in the model or set by the user
        Hint: Here would be good to set initial infection rate and perhaps track infection rate
        Also, make sure to update the def __init__  to include those properties.
        Ex. 
            def __init__(self, no_agents, width, height, infection_rate)
        """
        #CODE HERE
        self.infection_rate = infection_rate
        self.initial_infection = initial_infection
        self.recovery_time = recovery_time
        self.immunity_time = immunity_time
        self.mask_wearing_percentage = mask_wearing_percentage
        self.masked_infection_rate = masked_infection_rate
        self.death_rate = death_rate

        # Create agents and add it the schedule
        initially_infected = round(self.no_agents * self.initial_infection)
        for i in range(initially_infected): # Place initially infected agents
            a = ModifiedMaskAgent(i, self)
            a.infected = True
            a.recovery_time = recovery_time
            self.schedule.add(a)

            # Add the agent to a random grid cell
            x = self.random.randrange(self.grid.width)
            y = self.random.randrange(self.grid.height)
            self.grid.place_agent(a, (x, y))
        
        for i in range(no_agents, initially_infected, -1): # Place initially healthy agents
            a = ModifiedMaskAgent(i, self)
            self.schedule.add(a)

            # Add the agent to a random grid cell
            x = self.random.randrange(self.grid.width)
            y = self.random.randrange(self.grid.height)
            self.grid.place_agent(a, (x, y))
            
        agents = self.schedule.agents
        masked = round(mask_wearing_percentage * no_agents)
        agents_to_be_masked = random.sample(agents, masked)
        for agent in agents_to_be_masked:
            agent.masked = True
        
        """
        Add in to the datacollector properties you want to track over time in the graph.
        
        Ex. 
            self.datacollector = DataCollector({'Agents with Money': 'richAgents',....)
        """
        self.datacollector = mesa.DataCollector({"Infected Agents": "infectedAgents", "Immune Agents": "immuneAgents",
                                                 "Dead Agents": "deadAgents"})
        
    """
    Define any properties here that can be used for the DataCollector.
    Hint, number of infected agents, immune are pretty interesting.
    Ex. 
        @property
        def richAgents(self):
            agents = self.schedule.agents
            agents_with_money = [a.money_balance for a in agents]
            return int(np.sum(agents_with_money)))
    """
    @property
    def infectedAgents(self):
        agents = self.schedule.agents
        infected_agents = [a.infected for a in agents]
        return np.sum(infected_agents)
    
    @property
    def immuneAgents(self):
        agents = self.schedule.agents
        immune_agents = [a.immune for a in agents]
        return np.sum(immune_agents)
    
    @property
    def deadAgents(self):
        agents = self.schedule.agents
        dead_agents = [a.dead for a in agents]
        return np.sum(dead_agents)


    def step(self):
        self.datacollector.collect(self)
        self.schedule.step()
```

```python
"""
Task 10:
For this task, I created a "dead" variable. If an agent is dead, they do not move and they cannot infect.
This model is interesting because it allows us to see the disease "die off", often leaving the population with
a tragic amount of dead, but also ending any further spreading. This allows the model to better reflect real deadly
infectious diseases which might just end up killing everyone that could possibly be used as a disease vector.
"""
```
